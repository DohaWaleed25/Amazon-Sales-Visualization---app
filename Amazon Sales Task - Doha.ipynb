////// Data Preparartion & Cleaning /////// 

```python
# Change datatypes
df_clean = df.copy() # ÿπŸÖŸÑÿ™ ŸÜÿ≥ÿÆÿ© ÿ™ÿßŸÜŸäÿ© ŸÉŸÑŸäŸÜ ŸÖŸÜ ÿßŸÑÿØÿßÿ™ÿß Ÿà ÿ≥Ÿäÿ®ÿ™ ÿßŸÑÿßÿ≥ÿßÿ≥Ÿäÿ© ÿßÿ≠ÿ™Ÿäÿßÿ∑Ÿä 

for col in ['discounted_price' , 'actual_price']:
    df_clean[col] = (
        df_clean[col]
        .str.replace('‚Çπ','', regex = True)  # regex = Regular Expression ‚Üí ÿ£ÿØÿßÿ© ŸÇŸàŸäÿ© ŸÑŸÑÿ®ÿ≠ÿ´/ÿßŸÑÿßÿ≥ÿ™ÿ®ÿØÿßŸÑ ŸÅŸä ÿßŸÑŸÜÿµŸàÿµ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ "ÿ£ŸÜŸÖÿßÿ∑" ÿ®ÿØŸÑ ŸÖÿß ÿ™ŸÉŸàŸÜ ŸÉŸÑŸÖÿ© ÿ´ÿßÿ®ÿ™ÿ©
        .str.replace(',','', regex = True)
        .astype(float) 
    )

df_clean['discount_percentage'] = (
        df_clean['discount_percentage']
        .replace('%' , '', regex = True)
        .astype(float) 
    )

df_clean['rating'] = pd.to_numeric (df_clean['rating'], errors = 'coerce')
df_clean['rating_count'] = (
        df_clean['rating_count']
        .replace(',' , '' , regex = True)
        .astype(float)
)

print (df_clean.dtypes)
```

```python
print(df_clean.columns.tolist())
```

```python
# Drop unnecessary columns from clean data
df_clean = df_clean.drop(columns = ['img_link', 'product_link'], errors='ignore')
df_clean.head()
```

```python
# Check Dublicates 
print(df_clean.duplicated().sum())
```

```python
# Remove duplicate rows
df_clean = df_clean.drop_duplicates()
print(df_clean.duplicated().sum())
```

```python
print("\n Missing values per column (count):\n ", df_clean.isna().sum() , "\n")
```

```python
cat_cols = df_clean.select_dtypes(include=['object','category']).columns.tolist()
for c in cat_cols:
    print(c, "->", df_clean[c].nunique(), "unique; top:", df_clean[c].value_counts(dropna=False).head(5).to_dict())
```

```python
# Simple outlier screen for numeric columns
def iqr_outliers(series):
    q1, q3 = series.quantile([0.25,0.75])
    iqr = q3 - q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    return (series<lower) | (series>upper)

num_cols = df_clean.select_dtypes(include=[np.number]).columns
outlier_summary = {col: int(iqr_outliers(df_clean[col]).sum()) for col in num_cols if df_clean[col].notna().any()}
print("\nOutlier counts (IQR method):")
for k,v in outlier_summary.items():
    print(f"  {k}: {v}")
```

```python
# ÿπÿßŸàÿ≤ÿ© ÿßÿ¨Ÿäÿ® ÿßŸÑŸÅÿ±ŸÇ ÿπŸÑŸä ŸÖÿ≥ÿ™ŸàŸä ŸÉŸÑ ŸÇÿ∑ÿπÿ© ÿ®ŸäŸÜ ÿ≥ÿπÿ±Ÿáÿß ÿßŸÑÿßÿµŸÑŸä Ÿà ÿ≥ÿπÿ±Ÿáÿß ÿ®ÿπÿØ ÿßŸÑ discount 
df_clean['diff_of unit price'] = df_clean ['actual_price'] - df_clean['discounted_price']
```

```python
# ÿπÿßŸàÿ≤ÿ© ÿßÿ¨Ÿäÿ® ÿßŸÑÿπÿßÿ¶ÿØ ŸÖŸÜ ŸÉŸÑ ŸÇÿ∑ÿπÿ© ÿ®ÿπÿØ ÿßŸÑ discount
df_clean['revenue'] = df_clean['discounted_price'] * df_clean['rating_count']
```

- Q5: What are the top 5 products have the highest rating counts?

```python
df_clean['rating_count'] = (
    df_clean['rating_count']
    .astype(str)                 # ŸÜÿÆŸÑŸä ŸÉŸÑŸá string
    .str.replace(',', '')        # ŸÜÿ¥ŸäŸÑ ÿßŸÑŸÅŸàÿßÿµŸÑ
    .str.replace('\.0$', '', regex=True)  # ŸÜÿ¥ŸäŸÑ .0 ŸÖŸÜ ÿ¢ÿÆÿ± ÿßŸÑÿ±ŸÇŸÖ
    .replace('nan', '0')         # ŸÑŸà ŸÅŸä nan ŸÜÿÆŸÑŸäŸáÿß ÿµŸÅÿ±
    .astype(int)                 # ŸÜÿ≠ŸàŸÑŸá ŸÑ int
)

# ŸÜÿ¥ŸàŸÅ ÿ£ÿπŸÑŸâ 5 ŸÖŸÜÿ™ÿ¨ÿßÿ™ ÿ≠ÿ≥ÿ® ÿπÿØÿØ ÿßŸÑÿ±Ÿäÿ™ŸÜÿ¨
top_products = df_clean.nlargest(5, 'rating_count')
```

```python
# Top 5 Product with review
colors = sns.color_palette("Set2") 
plt.figure(figsize=(8,8))
plt.pie(
    top_products['rating_count'],
    labels=top_products['product_name'],
    autopct='%.2f%%',
    startangle=140,
    colors=colors
)
plt.title("Top 5 Most Reviewed Products")
plt.show()
```

- Q6: What are the top 5 products that had the highest discount?

```python
df_clean['discount_percentage'] = (
    df_clean['discount_percentage']
    .astype(str)           # ŸÜÿ≠ŸàŸÑŸá ŸÜÿµ
    .str.replace('%', '')  # ŸÜÿ¥ŸäŸÑ ÿπŸÑÿßŸÖÿ© %
    .astype(float)         # ŸÜÿ≠ŸàŸÑŸá ŸÑ float
)
```

```python
# ŸÜÿ¨Ÿäÿ® ÿ£ÿπŸÑŸâ 5 ŸÖŸÜÿ™ÿ¨ÿßÿ™ ÿ≠ÿ≥ÿ® ÿßŸÑÿÆÿµŸÖ
top_discount = df_clean.nlargest(5, 'discount_percentage')[['product_name', 'discount_percentage']]

print(top_discount)
```

```python
# ÿ£ŸàŸÑÿßŸã: ŸÜÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿßŸÑÿπŸÖŸàÿØ discount ÿ•ŸÜŸá ÿ±ŸÇŸÖŸä
df_clean['discount_percentage'] = (
    df_clean['discount_percentage']
    .astype(str)
    .str.replace('%', '', regex=False)  # ŸÜÿ¥ŸäŸÑ ÿπŸÑÿßŸÖÿ© %
    .astype(float)
)

# ŸÜÿ¨Ÿäÿ® ÿ£ÿπŸÑŸâ 5 ŸÖŸÜÿ™ÿ¨ÿßÿ™ ŸÑŸÉŸÑ Main Category
top_discount_by_category = (
    df_clean
    .sort_values(['main_category', 'discount_percentage'], ascending=[True, False])
    .groupby('main_category')
    .head(5)   # ÿ£ŸàŸÑ 5 ŸÖŸÜ ŸÉŸÑ ŸÉÿßÿ™Ÿäÿ¨Ÿàÿ±Ÿä
)

plt.figure(figsize=(14,8))
sns.barplot(
    data=top_discount_by_category,
    x='discount_percentage',
    y='product_name',
    hue='main_category',    # ŸÜŸÑŸàŸÜ ÿ≠ÿ≥ÿ® ÿßŸÑŸÉÿßÿ™Ÿäÿ¨Ÿàÿ±Ÿä
    dodge=False,
    palette="Set2"
)

plt.title("Top 5 Discounted Products per Main Category")
plt.xlabel("Discount (%)")
plt.ylabel("Product Name")
plt.legend(title="Main Category")
plt.show()
```

---> From each category, get highest discount product.
