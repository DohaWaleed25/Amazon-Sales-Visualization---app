////// Data Preparartion & Cleaning /////// 

```python
# Change datatypes
df_clean = df.copy() # عملت نسخة تانية كلين من الداتا و سيبت الاساسية احتياطي 

for col in ['discounted_price' , 'actual_price']:
    df_clean[col] = (
        df_clean[col]
        .str.replace('₹','', regex = True)  # regex = Regular Expression → أداة قوية للبحث/الاستبدال في النصوص باستخدام "أنماط" بدل ما تكون كلمة ثابتة
        .str.replace(',','', regex = True)
        .astype(float) 
    )

df_clean['discount_percentage'] = (
        df_clean['discount_percentage']
        .replace('%' , '', regex = True)
        .astype(float) 
    )

df_clean['rating'] = pd.to_numeric (df_clean['rating'], errors = 'coerce')
df_clean['rating_count'] = (
        df_clean['rating_count']
        .replace(',' , '' , regex = True)
        .astype(float)
)

print (df_clean.dtypes)
```

```python
print(df_clean.columns.tolist())
```

```python
# Drop unnecessary columns from clean data
df_clean = df_clean.drop(columns = ['img_link', 'product_link'], errors='ignore')
df_clean.head()
```

```python
# Check Dublicates 
print(df_clean.duplicated().sum())
```

```python
# Remove duplicate rows
df_clean = df_clean.drop_duplicates()
print(df_clean.duplicated().sum())
```

```python
print("\n Missing values per column (count):\n ", df_clean.isna().sum() , "\n")
```

--> There are 2 Null values at rating_count column & 1 Null value at rating after chande it to numeric. I will drop it. 

```python
# Drop null values 
df_clean = df_clean.dropna()
```

```python
print("\n Missing values per column (count):\n ", df_clean.isna().sum() , "\n") #Done
```

///// Discover أولي للأعمدة النصية (categorical columns) في الداتا

```python
cat_cols = df_clean.select_dtypes(include=['object','category']).columns.tolist()
for c in cat_cols:
    print(c, "->", df_clean[c].nunique(), "unique; top:", df_clean[c].value_counts(dropna=False).head(5).to_dict())
```

```python
# Simple outlier screen for numeric columns
def iqr_outliers(series):
    q1, q3 = series.quantile([0.25,0.75])
    iqr = q3 - q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    return (series<lower) | (series>upper)

num_cols = df_clean.select_dtypes(include=[np.number]).columns
outlier_summary = {col: int(iqr_outliers(df_clean[col]).sum()) for col in num_cols if df_clean[col].notna().any()}
print("\nOutlier counts (IQR method):")
for k,v in outlier_summary.items():
    print(f"  {k}: {v}")
```

```python
# عاوزة اجيب الفرق علي مستوي كل قطعة بين سعرها الاصلي و سعرها بعد ال discount 
df_clean['diff_of unit price'] = df_clean ['actual_price'] - df_clean['discounted_price']
```

```python
# عاوزة اجيب العائد من كل قطعة بعد ال discount
df_clean['revenue'] = df_clean['discounted_price'] * df_clean['rating_count']
```

```python
print(df_clean[['product_name','discounted_price', 'actual_price' , 'rating_count', 'diff_of unit price' , 'revenue']].head())
```

```python
# عاوزة اجيب التوب 20
top_products = df_clean.groupby('product_name')['revenue'].sum().nlargest(20).index
df_clean['Top20'] = df_clean['product_name'].isin(top_products)
df_top20 = df_clean[df_clean['product_name'].isin(top_products)]
```

- Q5: What are the top 5 products have the highest rating counts?

```python
df_clean['rating_count'] = (
    df_clean['rating_count']
    .astype(str)                 # نخلي كله string
    .str.replace(',', '')        # نشيل الفواصل
    .str.replace('\.0$', '', regex=True)  # نشيل .0 من آخر الرقم
    .replace('nan', '0')         # لو في nan نخليها صفر
    .astype(int)                 # نحوله ل int
)

# نشوف أعلى 5 منتجات حسب عدد الريتنج
top_products = df_clean.nlargest(5, 'rating_count')
```

```python
# Top 5 Product with review
colors = sns.color_palette("Set2") 
plt.figure(figsize=(8,8))
plt.pie(
    top_products['rating_count'],
    labels=top_products['product_name'],
    autopct='%.2f%%',
    startangle=140,
    colors=colors
)
plt.title("Top 5 Most Reviewed Products")
plt.show()
```

---> There are 4 products from total 5 have the same percentage and 1 has the lower one. 

- Q6: What are the top 5 products that had the highest discount?

```python
df_clean['discount_percentage'] = (
    df_clean['discount_percentage']
    .astype(str)           # نحوله نص
    .str.replace('%', '')  # نشيل علامة %
    .astype(float)         # نحوله ل float
)
```

```python
# نجيب أعلى 5 منتجات حسب الخصم
top_discount = df_clean.nlargest(5, 'discount_percentage')[['product_name', 'discount_percentage']]

print(top_discount)
```

```python
# أولاً: نتأكد من العمود discount إنه رقمي
df_clean['discount_percentage'] = (
    df_clean['discount_percentage']
    .astype(str)
    .str.replace('%', '', regex=False)  # نشيل علامة %
    .astype(float)
)

# نجيب أعلى 5 منتجات لكل Main Category
top_discount_by_category = (
    df_clean
    .sort_values(['main_category', 'discount_percentage'], ascending=[True, False])
    .groupby('main_category')
    .head(5)   # أول 5 من كل كاتيجوري
)

plt.figure(figsize=(14,8))
sns.barplot(
    data=top_discount_by_category,
    x='discount_percentage',
    y='product_name',
    hue='main_category',    # نلون حسب الكاتيجوري
    dodge=False,
    palette="Set2"
)

plt.title("Top 5 Discounted Products per Main Category")
plt.xlabel("Discount (%)")
plt.ylabel("Product Name")
plt.legend(title="Main Category")
plt.show()
```

---> From each category, get highest discount product. 

- Q7: What’s the average discount percentage across categories?
